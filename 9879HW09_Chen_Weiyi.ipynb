### 1. (12 points)
Following Almgren and Chriss,  assume that the stock price $S_t$ evolves as

$$
dS_t = \sigma\,dZ_t
$$

and the price $\tilde S_t$ at which we transact is given by

$$
\tilde S_t = S_t - \eta\,v_t
$$

where $v_t:=-{\dot x}_t$ is the rate of trading with $x_0 = X$ and $x_T = 0$.

In the lecture slides, we showed that with a risk term that penalizes average VaR instead of variance, the risk-adjusted cost of trading associated with a given price path $\{S_t\}$ is given by

$$
C = \int_0^T\,(S_t - \eta\,v_t) dx_t + \lambda\,\sigma\,\int_0^T\,x_t\,dt
$$

for some price of risk $\lambda$. We want to find a (possibly state-dependent) control $v_t$ that minimizes the expected cost $\cC=\E[C]$.

(a) Apply the technique of integration by parts to simplify the term $\int_0^T S_t dx_t$. Write down the HJB equation for the resulting problem.

<font color = 'darkblue'>

### Solution

Apply the technique of integration by parts,

$$
\left.\int_0^T S_t\,dx_t =  S_t\,x_t \right\vert_0^T - \int_0^T x_t\,dS_t = -XS -\int_0^T\sigma x_tdZ_t,
$$
we have
$$
\E\left[ \int_0^T S_t\,dx_t \right] = -XS
$$

Then the controlled $v_t$ that minimizes the exptected cost $\cC$ is

\begin{equation}
v_t 
= \mbox{argmin}_{v_t\in\mathcal G} \,\E\left[\cC\right] 
= \mbox{argmin}_{v_t\in\mathcal G} \left\{-XS + \E\left[\int_0^T\,\eta\,v_t^2 dt + \lambda\,\sigma\,\int_0^T\,x_t\,dt\right]\right\}
\end{equation}

Now the deterministic optimal control problem is to solve this part,
$$
\tilde C(t,x) = \min_{v_t\in\mathcal G} \left\{\E\left[\int_0^T\,\eta\,v_t^2 dt + \lambda\,\sigma\,\int_0^T\,x_t\,dt\right]\right\}
$$

with corresponding HJB equation as
$$
\frac{\partial{\tilde C}}{\partial t} + \lambda\sigma x_t + \min_{v_t\in\mathcal G} \left\{-\tilde C_x v + \eta v^2 \right\} = 0
$$

(b) Solve the first order condition to find the optimal trading rate $v^\star$.    

<font color='darkblue'>

### Solution

Let 
$$ \frac{\partial}{\partial v} (-\tilde C_x v + \eta v^2) = -\tilde C_x + 2\eta v = 0. $$

The optimal trading rate $v^\star$ is
$$ v^\star = \frac{1}{2\eta}\tilde C_x. $$

(c) Substitute $v^\star$ back into the HJB equation to show that the value function $\cC$ must satisfy
$$
\cC_t = \frac{\cC_x^2}{4\,\eta\,} - \lambda \sigma x.
$$

<font color='darkblue'>

### Solution

Substitute $v^\star$ back to HJB,
\begin{equation}
    \frac{\partial{C}}{\partial t} + \lambda\sigma x_t + \left\{-C_x \cdot \frac{1}{2\eta}C_x + \eta \cdot \frac{1}{4\eta^2}C_x^2 \right\} = 0
\end{equation}

Obviously,

\begin{equation}
    \cC_t = \frac{\cC_x^2}{4\,\eta\,} - \lambda \sigma x.
\end{equation}


(d) With the terminal condition, 
$$\lim_{t\uparrow T}\cC(t,x) = \begin{cases}0&\text{if $x=0$,}\\
+\infty&\text{if $x\neq0$.}
\end{cases}$$
solve the HJB equaiton with the ansatz $\cC(t,x) = a(t)x^2 + b(t)x + c(t)$. Verify that the optimal trading strategy with liquidation horizon set to the characteristic time is given by
\begin{eqnarray*}
x_t&=&X\,\left(1-\frac{t}{T}\right)^2.
\end{eqnarray*}
Recall that with the notation of the lecture slides, the characteristic time is defined as
$$
T^\star=\sqrt{\frac{2\,X}{A}}=\sqrt{\frac{4\,\eta\,X}{\lambda\,\sigma}}
$$
Compute the risk-adjusted cost function $\cC$ associated with this strategy. 

<font color='darkblue'>

### Solution

In the lecture notes, we have
$$ x_t = X\left( 1 - \frac{t}{T}\right)^2, v_t = -\dot{x}_t = \frac{2X}{T}\left( 1 - \frac{t}{T}\right) $$

By substituting them into the cost function, we finally derive
\begin{equation}
\cC(t,x) 
= \int_t^T \lambda\sigma\,x_s\,ds+\lambda\sigma\int_t^T x_s\,ds 
= \frac{2\lambda\sigma T }{3X^{\frac{1}{2}}}x_t^{\frac{3}{2}}
\end{equation}

(a) The *msft.bats* dataset from *MSFT130311.rData* contains all trades on the BATS exchange on 11-Mar-2013.  One trading days has 390 minutes.  Sample the  *msft.bats* dataset evenly, roughly every minute of volume time.  Superimpose plots of the optimal strategy under ABM and GBM assuming liquidation takes place over one trading day.

<font color='darkblue'>

### Solution

```python
%%R

# initialization

df <- msft130311.bats    # The msft.bats dataset from MSFT130311.rData
i_T <- 390               # One trading days has 390 minutes, liquidation time T
i_X <- 10000             # initial positions
ls_timeStamp <- c(1:i_T)   # list of time stamps to use later

# prepare metrics

df$sum <- cumsum(abs(df$signed.shares))     # cumulative sum of shares number
i_numTrade <- length(df$sum)                # total number of trades
i_totalVol <- df$sum[i_numTrade]              # total volumn

# sample the msft.bats dataset evenly, roughly every minute of volume time

ls_samples <- seq(1, i_totalVol, i_totalVol/i_T)

# construct a list of sample index

ls_index <- c(1:length(ls_samples))
j = 1
for (i in 1:i_numTrade) {
    if (df$sum[i] > ls_samples[j] && j <=390) {
        ls_index[j] = i
        j = j+1
    }
}

# construct price series S_t, position list and trading rate list
ts_price <- df[ls_index, 'trade.price']

ls_xABM <- c(0:i_T)
ls_rABM <- c(1:i_T) 
ls_xABM[1] = i_X

ls_xGBM <- c(0:i_T)
ls_rGBM <- c(1:i_T) 
ls_xGBM[1] = i_X


# calculate trading rate and positions

for (t in 1:i_T){
    ls_rABM[t] = abs(ls_xABM[t])/(i_T - t) + i_X / i_T * (1 - t/i_T)
    ls_xABM[t+1] = max(ls_xABM[t] - ls_rABM[t], 0)
    
    ls_rGBM[t] = abs(ls_xGBM[t])/(i_T - t) + i_X / i_T * ts_price[t]/ts_price[1] * (1 - t/i_T)
    ls_xGBM[t+1] = max(ls_xGBM[t] - ls_rGBM[t], 0)
}

# plot the trading rate
plot(ls_timeStamp, ls_rABM, xlab = "Time Stamps", ylab = "Trading Rates", col = "red", type = "p", main = 'ABM')
plot(ls_timeStamp, ls_rGBM, col = "blue", type = "l", lwd = 3, main='GBM')

plot(ls_timeStamp, ls_rABM, xlab = "Time Stamps", ylab = "Trading Rates", col = "red", type = "p", main = 'ABM & GBM')
lines(ls_timeStamp, ls_rGBM, col = "blue", type = "l", lwd = 3)
legend("topright", c("ABM","GBM"), col = c("red","blue"), lwd = 2, inset = 0.03)
```

## Kratz and Schöneborn with time-averaged VaR risk charge

### 3. (16 points)

(a) Repeat the computation of the optimal strategy from the lecture slides with time-averaged VaR

$$
\lambda\,\sigma\,\int_0^T\,x_t\,dt
$$

as the risk term (rather than the quadratic variation $\lambda\,\sigma^2\,\int_0^T\,x_t^2\,dt$).  Show that the optimal position

$$
x^\star(t)=\left(X-\frac{A\, T}{\theta }\right)\,\frac{1-e^{-\theta\,(T-t)} }{1-e^{-\theta\,T }}+\frac{A }{\theta }\,(T-t)
$$
and deduce the optimal trading rate $v^\star(t)$.

<font color='darkblue'>

### Solution

As before, the expected cost of liquidation (with risk penalty) is given
(as of time $t$) by

$$\mathcal{C} (t)=\eta\,\mathbb{E} \left[\int_t^T\, (\dot x_u^2+\kappa^2\,x_u^2)\,du\right]$$

with $$\kappa^2=\frac{\lambda\,\sigma^2}{\eta}.$$

In our case,

\begin{eqnarray*}
\mathcal{C} = \eta\,\E\left[\,\int_0^T\,({\dot x}_t^2 + \frac{\lambda\sigma}{\eta}x_t) 1_{t < t*}\,dt\right]
= \eta\,\int_0^T\,({\dot x}_t^2 + \frac{\lambda\sigma}{\eta}x_t) e^{-\theta t}\,dt
\end{eqnarray*}

Apply Kratz-Schöneborn HJB equation, we derive 

$$
\frac{d}{dt}(2\dot{x}_te^{-\theta t}) - \frac{\lambda\sigma}{\eta}e^{-\theta t} = 0 \Rightarrow\,\, \ddot{x}_t - \theta\dot{x}_t - \frac{\lambda\sigma}{2\eta} = 0
$$

Given the boundary conditions as $X_0 = X$, $X_T = 0$ and denote $A = \frac{\lambda\sigma}{2\eta}$, we have the solution of the ODE

$$ x^\star(t)=\left(X-\frac{A\, T}{\theta }\right)\,\frac{1-e^{-\theta\,(T-t)} }{1-e^{-\theta\,T }}+\frac{A }{\theta }\,(T-t) $$

The optimal trading rate is
$$
v^\star(t)= -\dot x^\star(t) = \theta(X-\frac{AT}{\theta})\frac{e^{-\theta(T-t)}}{1-e^{-\theta T}} + \frac{A}{\theta}.
$$

(b) By Taylor-expanding the solution around $\theta=0$, show that we retrieve the solution derived in the slides with no dark pool

$$
x^\star(t)=\left(X-\frac{A \, T}{2}\,t\right)\,\left(1-\frac{t}{T}\right) .
$$

in the limit $\theta \to 0$.

<font color = "darkblue"> 

### Solution

By Taylor-expanding the solution around $\theta = 0$,
\begin{eqnarray*}
x^\star(t) &\approx& \left(X-\frac{A\, T}{\theta }\right)\,\frac{(T-t)\theta-\frac{1}{2}(T-t)^2\theta^2}{T\theta - \frac{1}{2}T^2\theta^2}+\frac{A }{\theta }\,(T-t)\\
&=& \left(X-\frac{A\, T}{\theta }\right)\,\frac{(T-t)-\frac{1}{2}(T-t)^2\theta}{T - \frac{1}{2}T^2\theta}+\frac{A }{\theta }\,(T-t)\\
\end{eqnarray*}

For the part with $X$,

\begin{eqnarray}
X \frac{(T-t)-\frac{1}{2}(T-t)^2\theta}{T - \frac{1}{2}T^2\theta} & \approx & X\frac{T-t}{T} = X(1-\frac{t}{T})
\end{eqnarray}

For the others,

\begin{eqnarray}
-\frac{AT}{\theta} \frac{(T-t)-\frac{1}{2}(T-t)^2\theta}{T - \frac{1}{2}T^2\theta} +\frac{A }{\theta }\,(T-t)
&=& - A(T-t)\left[\frac{1-\frac{1}{2}(T-t)\theta}{\theta(1-\frac{1}{2}T\theta)} - \frac{1}{\theta}\right] \\
&=& - A(T-t)\left[\frac{\frac{1}{2}t}{1-\frac{1}{2}T\theta} \right]\\
&\approx& - \frac{A(T-t)t}{2}\\
&=& \left(-\frac{AT}{2}t \right)\,\left(1-\frac{t}{T} \right)
\end{eqnarray}

Therefore aggregating together,

$$
x^\star(t)=\left(X-\frac{A \, T}{2}\,t\right)\,\left(1-\frac{t}{T}\right) .
$$
