Crash of May 6, 2010? Why did it cause a crash? How could this algorithm have been
amended so as to avoid the ensuing disaster?

#### Answer

The algorithmic trading strategy being used by the seller who triggered the Flash Crash of May 6, 2010 is constant participation rate. 

He splitted futures contracts selling orders to each "child" order, as of a fixed fraction of the total trading volume, for the purpose of reducing market impact. But the "child" order would become larger and larger as the "parent" order gets larger, the selling orders quickly exhausted market's liquidity and the price dropped sharply. 

To avoid the ensuing disaster, just let the algorithm have some stop-loss exits. The algo should detect the price, and once it drops sharply, it should instantly stopped until the liquidity of market resumes.

### 3. (16 points)

(a) Using the code supplied to generate Figure 2 as a template, simulate 100,000 events, computing the average book shape (in a band $\pm 20$ around the mid price) for the
following sets of parameters:

|$\alpha$|$\mu$|$\delta$|
-----|-----:|-------:|
I| 1 | 10 | $\frac15$
II| 1 | 8 |  $\frac15$
III| 1 |  10 | $\frac16$
IV| 1  | 10 | $\frac18$ 

(Warning: This might take longer than 15 minutes to run!).

(b) Generate a summary plot of the average book shapes for each parameter set (I, II, II, and IV).

(c) Verify that the relationships between the slopes of the order books at-the-money for each parameter set are consistent with the predictions from dimensional analysis.

(d) Verify that the asymptotic book depths are consistent with the predictions from dimensional analysis.

```python
%%R

#### (a) & (b)

# define function

compute_averge_book_shape = function(alpha, mu, delta) {
    
    # parameters setting
    
    alpha = alpha
    mu = mu
    delta = delta
    
    # initialize the order book
    
    initializeBook5()

    # Burn in for 1000 events

    for(count in 1:1000){
        generateEvent()
    }

    # Simulate 100,000 events

    numEvents = 100000
    avgBookShape = bookShape(20)/numEvents
    for(count in 2:numEvents){
        generateEvent()
        avgBookShape <- avgBookShape+bookShape(20)/numEvents
    }
    
    # plot the resulting book shape
    
    plot(-20:20,avgBookShape,main="Parameter set",xlab="Relative price",ylab="# Shares", col="red", type="b")
}

# 1st set of parameters 

compute_averge_book_shape(1, 10, 1/5)

# 2nd set of parameters 

compute_averge_book_shape(1, 8, 1/5)

# 3rd set of parameters 

compute_averge_book_shape(1, 8, 1/6)

# 4th set of parameters 

compute_averge_book_shape(1, 10, 1/8)
```

#### (c)

The slopes of the order book are 0.5, 0.625, 0.6, 0.8 respectively, given by theoretical formula:
    $$ \frac{\alpha^2}{\mu\,\delta} $$
Though the slopes of the graphs in part (b) are a bit larger than the theoretical values, they are in right order and same scales. Therefore, the relationships between the slopes of the order books at-the-money for each parameter set are consistent with the predictions from dimensional analysis.

#### (d)

The asymptotic book depths of the order book are 5, 5, 6, 8 respectively, given by theoretical formula:
    $$
    \alpha/\delta
    $$
which are consistent with the predictions from dimensional analysis.
